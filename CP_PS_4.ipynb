## Problem 1:

Plot the critical supersaturation and critical radius as a function of the radius of a dry $NaCl$ particle.  The range of $NaCl$ sizes should cover the sizes that you would expect to find in the atmosphere.  If you need to make additional assumptions about temperatire, pressure or any other atmospheric parameters, use values that are close to those you would expect to find in the atmosphere.  Repeat the calculation and plot for $(NH_4)_2(SO_4)$.  

According to the notes, the dry radius of $NaCl$ varies from .02 to .10 microns in the real atmosphere.  Similarly the dry radius of $(NH_4)_2(SO_4)$ varies from .02 microns to .05 microns.  Masses vary from $10^{-19}kg$ to $10^{-17}kg$ for $NaCl$ and $10^{-19}kg$ and $10^{-18}kg$ for $(NH_4)_2(SO_4)$.  

```python
r_nacl = np.linspace(.02, .10, 1000)/1e6 #meters
r_salt2 = np.linspace(.02,.05,1000)/1e6 #meters
```

Equations 6.7 and 6.8 in Rogers and Yao give expressions for $r^*$ and $S^*$:

$r^* = \sqrt{3b/a}$

$S^* = 1+\sqrt{4a^3/27b}$

where:

$a = \frac{2\sigma}{\rho R_v} \approx \frac{5.5e-3}{T}$ in units of $cm$

$b = \frac{i r_{salt}^3 \rho_{salt} M_{water}}{\rho_{water} M_{salt}}$ in units of $cm^3$ 


```python
T=273 #K

i=2 #suggested in Rogers and Yao

rho_nacl = 2.16 #gcm^-3
rho_nacl = rho_nacl/1000 #kg cm^-3
rho_nacl = rho_nacl*1e-6 #kg m^-3

rho_water = 1000 #kgm^-3

rho_salt2 = 1.77 #gcm^-3
rho_salt2 = rho_salt2/1000 #kg/cm^-3
rho_salt2 = rho_salt2*1e-6 #kgm^-3

mass_water = 18.01528 #amu
mass_nacl = 58.44277 #amu
mass_salt2 = 132.09356 #amu

a = (3.3e-5/T)/100 #meters
b_nacl = (i*(r_nacl**3)*rho_nacl*mass_water)/(rho_water*mass_nacl) #meters^3
b_salt2 = (i*(r_salt2**3)*rho_salt2*mass_water)/(rho_water*mass_salt2) #meters^3
```

```python
r_star_nacl = np.sqrt(3*b_nacl/a)
S_star_nacl = 1+np.sqrt(4*a**3/(27*b_nacl))

r_star_salt2 = np.sqrt(3*b_salt2/a)
S_star_salt2 = 1+np.sqrt(4*a**3/(27*b_salt2))
```

```python
plt.plot(r_nacl*1e6,S_star_nacl,label='NaCl')
plt.plot(r_salt2*1e6,S_star_salt2,label='[(NH4)_2][SO4]')
plt.grid(visible='True',axis='both')
plt.legend()
plt.title("Critical Saturation vs. Salt Radius")
plt.xlabel('Salt Radius (microns)')
plt.ylabel('Critical Saturation (%)')
```

```python

plt.plot(r_nacl*1e6, r_star_nacl*1e6,label='NaCl') 
plt.plot(r_salt2*1e6, r_star_salt2*1e6,label='[(NH4)_2][SO4]')
plt.grid(visible='True',axis='both')
plt.legend()
plt.title("Critical Radius Growth Radius vs. Salt Radius")
plt.xlabel('Salt Radius (microns)')
plt.ylabel('Critical Growth Radius (microns)')
```
